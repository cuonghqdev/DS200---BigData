```python
import os
import sys

```

# Tạo hàm làm sạch

```python
def clean_file(input_path, output_path):
    print(f"Đang xử lý: {input_path} -> {output_path}...")
    try:
        with open(input_path, 'r', encoding='utf-8') as f_in, open(output_path, 'w', encoding='utf-8') as f_out:
            buffer = ''
            for line in f_in:
                line = line.strip()
                if not line:
                    continue
                # Nối dòng hiện tại vào buffer nếu buffer đang có dữ liệu
                if buffer:
                    potential_line = buffer + line
                else:
                    potential_line = line

                # Đếm số dấu phẩy để xác định dòng có hoàn chỉnh hay không
                # ratings: 3 dấu phẩy, users: 4 dấu phẩy, movies: 2 dấu phẩy
                comma_count = potential_line.count(',')
                is_complete = (('ratings' in os.path.basename(input_path) and comma_count == 3) or
                               ('users' in os.path.basename(input_path) and comma_count == 4) or
                               ('movies' in os.path.basename(input_path) and comma_count == 2))

                if is_complete:
                    f_out.write(potential_line + '\n')
                    buffer = '' 
                else:
                    buffer = potential_line
        print("Hoàn thành.")
    except FileNotFoundError:
        print(f"!!! LỖI: Không tìm thấy file tại đường dẫn: {input_path}", file=sys.stderr)
    except Exception as e:
        print(f"!!! LỖI không xác định khi xử lý file {input_path}: {e}", file=sys.stderr)


```

# Thực thi hàm 

```python

base_path = r'G:\.shortcut-targets-by-id\1eDKYURr-Qm222Ul6PoZLc4b99cuwHTC3\Cường\Đại học\DS200\Lab 1'

files_to_clean = ['ratings_1.txt', 'ratings_2.txt', 'users.txt', 'movies.txt']

# Chạy vòng lặp để làm sạch từng file
for filename in files_to_clean:
    input_file_path = os.path.join(base_path, filename)
    output_file_path = os.path.join(base_path, filename.replace('.txt', '.cleaned.txt'))
    clean_file(input_file_path, output_file_path)
```

# Gộp 2 file ratings đã được làm sạch

```python
print("\nBắt đầu gộp các file ratings...")
try:
    ratings1_cleaned = os.path.join(base_path, 'ratings_1.cleaned.txt')
    ratings2_cleaned = os.path.join(base_path, 'ratings_2.cleaned.txt')
    combined_ratings_file = os.path.join(base_path, 'ratings.cleaned.txt')

    with open(combined_ratings_file, 'w', encoding='utf-8') as outfile:
        for fname in [ratings1_cleaned, ratings2_cleaned]:
            with open(fname, 'r', encoding='utf-8') as infile:
                outfile.write(infile.read())
    print(f"Đã gộp thành công và lưu tại: {combined_ratings_file}")
except FileNotFoundError as e:
    print(f"!!! LỖI khi gộp file: Không tìm thấy file cần thiết. Vui lòng kiểm tra lại. {e}", file=sys.stderr)
except Exception as e:
    print(f"!!! LỖI không xác định khi gộp file: {e}", file=sys.stderr)
```
